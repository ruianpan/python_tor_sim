```python
import socket
import sys
from relayInfo import relay
def recvall(sock):
    BUFF_SIZE = 4096 # 4
    data = b''
    while True:
        part = sock.recv(BUFF_SIZE)
        data += part
        if len(part) < BUFF_SIZE:
            break
    return data

def sendrelayInfoList(relayInfoList,connection):
    MSG = ""
    for relayInfo in relayInfoList:
        MSG+='#'
        MSG+=str(relayInfo.relayId)
        for i,j in zip(relayInfo.circuitIdList, relayInfo.circuitBandList):
            MSG = MSG +'I' +str(i)+ 'B'+str(j)
    connection.send(MSG)
            
        



sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
server_address = ('localhost', 3490)
sock.bind(server_address)
sock.listen(1)
relayList = []
relayInfoList = []
while True: 
    connection, client_address = sock.accept()
    recvdata = recvall(connection)
    
    
    
    
    msg = recvdata.decode('utf-8')
    
    if msg[0] == 'C':
        sendrelayInfoList(relayInfoList,connection)
        continue
    
    connection.close()
    
    currRelay = relay()
    stringToParse = str(msg)
    relayIndex = stringToParse.find('#')
    relayId = int(stringToParse[relayIndex+1:])
    currRelay.circuitsFromString(stringToParse[:relayIndex])
    currRelay.relayId = relayId
    if relayId not in relayList:
        print "relay info received"
        relayInfoList.append(currRelay)
        relayList.append(relayId)
    else:
        print "relay info updated"
        for oldrelay in relayInfoList:
            if oldrelay.relayId == currRelay.relayId:
                relayInfoList.remove(oldrelay)
                relayInfoList.append(currRelay)
                
  
```
